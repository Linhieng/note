## ğŸ• loc å’Œ iloc

`loc`: location. é€šè¿‡æ ‡ç­¾æ¥æå–å¯¹åº”åˆ—ã€‚

`iloc`: integer location, é€šè¿‡æ•´æ•°ç´¢å¼•(è¡Œå·,åˆ—å·)æ¥æå–å¯¹åº”è¡Œåˆ—

ä»–ä»¬éƒ½æ¥æ”¶ä¸¤ä¸ªå‚æ•°, ç¬¬ä¸€ä¸ªå‚æ•°ç”¨äºæŒ‡å®šè¡Œ, ç¬¬äºŒä¸ªå‚æ•°ç”¨äºæŒ‡å®šåˆ—

ç›´æ¥é€šè¿‡ `[columns]` æ–¹å¼, æ˜¯ç›¸å½“äº `loc[:, columns]`, å³ç›´æ¥ç´¢å¼•æ—¶, `columns` ä¼šè¢«è®¤ä¸ºæ˜¯åˆ—å

å¦‚æœåŒæ—¶ä½¿ç”¨æ•°å­—å’Œå­—ç¬¦ä¸²è¿›è¡Œå®šä½, è¯·ä½¿ç”¨ `loc` è€Œä¸æ˜¯ `iloc`ã€‚

`loc` å’Œ `iloc` éƒ½å¯ä»¥ä½¿ç”¨æ•°å­—, ä»–ä»¬çš„åŒºåˆ«åœ¨äº, `loc` ä¸­çš„"æ•°å­—"æ˜¯æ ‡ç­¾, å³ç´¢å¼•å€¼, è€Œ `iloc` ä¸­çš„æ•°å­—æ˜¯è¡Œå·,åˆ—å·ã€‚å…·ä½“è¯·è§ä¸‹é¢è¿™ä¸ªä¾‹å­:

````python
import pandas as pd

# å°†æ•°ç»„è½¬æ¢ä¸ºæ•°æ®æ¡†æ¶
df = pd.DataFrame([i for i in range(10)]).sample(frac=1)
print(df.iloc[[1,2,3], :]) # å–å‰ä¸‰è¡Œ, æ­¤æ—¶å‰ä¸‰è¡Œçš„ç´¢å¼•ä¸ä¸€å®šæ˜¯ 1,2,3
print(df.loc[[1,2,3], :]) # å–ä¸‰è¡Œ, æŒ‡å®šä»–ä»¬çš„ç´¢å¼•ä¸º 1,2,3
````

## 1. ä½¿ç”¨ `loc` å¿«æ·æ›¿æ¢æŒ‡å®šè¡ŒæŒ‡å®šåˆ—çš„å†…å®¹

````python
import pandas as pd

# åˆ›å»ºä¸€ä¸ªDataFrame
df = pd.DataFrame({
    'c': [1],
    'score-1': [10.9],
    'score-2': [10.9],
    'score-3': [10.9],
    'score-4': [10.93],
    'score-5': [10.93],
    'Mean recall scores': [1]
})

print(df)


# è¦æ›¿æ¢çš„è¡Œå’Œåˆ—
index = 0
columns =  [f'score-{i+1}' for i in range(5)]

# è¦æ›¿æ¢æˆçš„æ–°å€¼
scores = [9, 9, 9, 93, 93]
data = {'score-' + str(i+1): scores[i] for i in range(len(scores))}
new_row = pd.Series(data)

# ä½¿ç”¨ loc å‡½æ•°è¿›è¡Œæ›¿æ¢
df.loc[index, columns] = new_row


print(df)
````

## 2. ä½¿ç”¨ `concat` æ–°å¢è¡Œæ•°æ®

````python
import pandas as pd
import random as rd

# åˆ›å»ºä¸€ä¸ªåŒ…å«æ•°å€¼çš„æ•°ç»„
scores = [rd.randint(0,10) for i in range(len(scores))]
# ä½¿ç”¨å­—å…¸æ¨å¯¼å¼åˆ›å»ºå­—å…¸
data = {'score-' + str(i+1): scores[i] for i in range(len(scores))}
row = pd.Series(data)

K = 5
o = ['C_parameter', 'Mean recall scores']
n = [f'score-{i}' for i in range(1, K+1)]
columns = o[:1] + n + o[1:]
df = pd.DataFrame(columns=columns)

# row.to_frame() å°† Series è½¬æ¢ä¸ºä¸€ä¸ª DataFrame å¯¹è±¡
df = pd.concat([df, row.to_frame().T], ignore_index=True)
df.loc[0, 'C_parameter'] = 1
df.loc[0, 'Mean recall scores'] = 1

print(df)

scores = [rd.randint(0,10) for i in range(len(scores))]
data = {'score-' + str(i+1): scores[i] for i in range(len(scores))}
row = pd.Series(data)
df = pd.concat([df, row.to_frame().T], ignore_index=False)

print(df)

df.loc[1, 'C_parameter'] = 2
df.loc[1, 'Mean recall scores'] = 2

print(df)

scores = [rd.randint(0,10) for i in range(len(scores))]
data = {'score-' + str(i+1): scores[i] for i in range(len(scores))}
row = pd.Series(data)
df = pd.concat([df, row.to_frame().T], ignore_index=True)

print(df)
````

##  è·å– A åˆ—çš„æœ€å¤§å€¼æ‰€åœ¨è¡Œ B åˆ—çš„å€¼

````python
import numpy as np
import pandas as pd

# éšæœºç”Ÿæˆ 5x5 æ•°ç»„
np.random.seed(42)
data = np.random.rand(5, 5)

# å°†æ•°ç»„è½¬æ¢ä¸ºæ•°æ®æ¡†æ¶
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])

# æ˜¾ç¤ºæ•°æ®æ¡†æ¶
print(df)
print('')

# æ–¹æ³•1
i_max = df['A'].idxmax()
print(i_max)
maxA_B = df.loc[i_max, 'B']
print(maxA_B)
print('')

# æ–¹æ³•2
maxC = df['C'].max()
print(maxC)
row_with_max_c = df.loc[df['C'] == maxC]
print(row_with_max_c)
maxC_D = row_with_max_c.iloc[0]['D']
print(maxC_D)
````

## pd.cut æ¡ˆä¾‹

````python
import pandas as pd
import numpy as np

# éšæœºç”Ÿæˆä¸€äº›æ•°æ®
df = pd.DataFrame({"A": np.random.randint(0, 100, 10)})

# å®šä¹‰åŒºé—´èŒƒå›´
bins = [0, 20, 40, 60, 80, 100]

# ä½¿ç”¨ pd.cut è¿›è¡ŒåŒºé—´åˆ’åˆ†
df["category"] = pd.cut(df["A"], bins)

# æ‰“å°ç»“æœ
print(df)
````

````python
import pandas as pd
import numpy as np

# éšæœºç”Ÿæˆä¸€äº›æ•°æ®
df = pd.DataFrame({"A": np.random.randint(-100, 100, 10)})

# å®šä¹‰åŒºé—´èŒƒå›´
bins = [-100, 0, 100]
labels = ['è´Ÿæ•°-', 'æ­£æ•°+']

# ä½¿ç”¨ pd.cut è¿›è¡ŒåŒºé—´åˆ’åˆ†
df["category"] = pd.cut(df["A"], bins=bins, labels=labels)

# æ‰“å°ç»“æœ
print(df)
````
